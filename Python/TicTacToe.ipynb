```python
import random
random.randint(1,2) #returns a random number between 1 and 2 both inclusive
```

```python
def playerToss():
    if random.randint(1,2)==1:
        return 'player1'
    else:
        return 'player2'
        
```

```python
def choose_marker(turn):
    marker=' '
    while marker not in ['X','O']:
        marker= input("choose 'X' or 'O'")
    return marker
```

```python
def getBoard():
    return [' ']*10
```

```python
getBoard()
```

```python
def space_check(board, position):
    
    return board[position] == ' '
```

```python
def player_choice(board,turn):
    position = 0
    
    while position not in [1,2,3,4,5,6,7,8,9] or not space_check(board, position):
        position = int(input('{}Choose your next position: (1-9) '.format(turn)))
        
    return position
```

```python
def place_marker(board, markers, position,turn):
    board[position]=markers[turn]
```

```python
def display_board(board):
 
    print('   |   |')
    print(' ' + board[7] + ' | ' + board[8] + ' | ' + board[9])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[4] + ' | ' + board[5] + ' | ' + board[6])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[1] + ' | ' + board[2] + ' | ' + board[3])
    print('   |   |')
```

```python
def win_check(board,turn,markers):
    
    mark=markers[turn]
    ''' check rows 
         board[1]==board[2]==board[3]==mark
    check columns
    check diagonals
    
    '''
    
    return ((board[7] == mark and board[8] == mark and board[9] == mark) or # across the top
    (board[4] == mark and board[5] == mark and board[6] == mark) or # across the middle
    (board[1] == mark and board[2] == mark and board[3] == mark) or # across the bottom
    (board[7] == mark and board[4] == mark and board[1] == mark) or # down the middle
    (board[8] == mark and board[5] == mark and board[2] == mark) or # down the middle
    (board[9] == mark and board[6] == mark and board[3] == mark) or # down the right side
    (board[7] == mark and board[5] == mark and board[3] == mark) or # diagonal
    (board[9] == mark and board[5] == mark and board[1] == mark)) # diagonal
```

```python
def full_board_check(board):
    for i in range(1,10):
        if space_check(board, i):
            return False
    return True
```

```python
def restart():
    x=input('Do you want to restart Y/N')
    return (x=='Y')
    
```

```python
def swapPlayers(turn):
    if (turn=='player1'):
        return 'player2'
    else:
        return 'player1'
```

```python
def TIC_TAC_Toe():
    turn=None
    game_on=True
    print("hi")
    while game_on:
        board=getBoard()
        display_board(board)
        turn=playerToss()
        print("{0} won the toss".format(turn))
        marker  = choose_marker(turn)
        if turn == 'player1':
            if marker=='X':
                markers={'player1':'X','player2':'O'}
            else:
                markers ={'player1':'O','player2':'X'}
        else:
            if marker=='X':
                markers={'player2':'X','player1':'O'}
            else:
                markers={'player2':'O','player1':'X'}
        while True:
            pos=player_choice(board,turn)
            place_marker(board,markers,pos,turn)
            display_board(board)
            if win_check(board,turn,markers):
                print("{}has won the game".format(turn))
                if(not restart()):
                    game_on=False
                break
            elif(full_board_check(board)):
                print("draw")
                if(not restart()):
                    game_on=False
                break
            else:
                turn=swapPlayers(turn)
        
        
        
    
```

```python
TIC_TAC_Toe()
```
